# PUNTO 2)

**RICOSTRUZIONE NUMERICA DELLA FORMA D'ONDA A PINNA DI SQUALO IN USCITA DA INTEGRATORE, CON FORMA QUADRA IN INGRESSO, IN FUNZIONE DELLA FREQUENZA (PER UNA DATA FREQUENZA DI TAGLIO)**

```python
from array import array
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square

n = 500
fT= 482
#y = 0.5*square(x)
def run(f):
  x = np.linspace(-2/f,2/f,5000)
  somma = np.zeros_like(x)
  for i in range(n):
    fk = (2*i+1) * f
    function = 2/((2*i+1)*np.pi) /np.sqrt(1+(fk/fT)**2) * np.sin(2*np.pi*fk * x - np.arctan(fk/fT))
    somma += function
  return [x,somma]

fig = plt.figure('Ricostruzione pinna')

fig.add_axes((0.075, 0.690, 0.4, 0.25))
plt.title('Ricostruzione $f=25$ Hz')
a = run(25)
plt.plot(a[0],a[1], label='Simulazione n=5')

fig.add_axes((0.575, 0.690, 0.4, 0.25))
plt.title('Ricostruzione $f=50$ Hz')
a = run(50)
plt.plot(a[0],a[1], label='Simulazione n=5')

fig.add_axes((0.075, 0.360, 0.4, 0.25))
plt.title('Ricostruzione $f=250$ Hz')
plt.ylabel("Ampiezza [arb.un.]")
a = run(250)
plt.plot(a[0],a[1], label='Simulazione n=5')

fig.add_axes((0.575, 0.360, 0.4, 0.25))
plt.title('Ricostruzione $f=500$ Hz')
plt.ylabel("Ampiezza [arb.un.]")
a = run(500)
plt.plot(a[0],a[1], label='Simulazione n=5')

fig.add_axes((0.075, 0.045, 0.4, 0.25))
plt.title('Ricostruzione $f=2500$ Hz')
plt.xlabel("Tempo [s]")
a = run(2500)
plt.plot(a[0],a[1], label='Simulazione n=5')

fig.add_axes((0.575, 0.045, 0.4, 0.25))
plt.title('Ricostruzione $f=5000$ Hz')
plt.xlabel("Tempo [s]")
a = run(5000)
plt.plot(a[0],a[1], label='Simulazione n=5')
plt.show()
```

# PUNTO 3)


**SIMULAZIONE NUMERICA DEI SEGNALI A PINNA DI SQUALO ACQUISITI DA ARDUINO**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

colonne = ['t', 't_err', 'V', 'V_err']
df = pd.read_table("/content/data/esp6-1.0082kHz.txt", skiprows=0, sep='\s+', header=None, names=colonne)
t = (df['t'] - 475.82)/1e6
V = df['V']
sigma_V = df['V_err']

x = np.linspace(0,np.max(t),1000)
def curve(x, f, A, C, phi):
  n = 500
  fT= 482
  somma = C
  for i in range(n):
    fk = (2*i+1) * f
    function = 2*A/((2*i+1)*np.pi) /np.sqrt(1+(fk/fT)**2) * np.sin(2*np.pi*fk * x - np.arctan(fk/fT) + phi)
    somma += function
  return somma

popt, pcov = curve_fit(curve, t, V, sigma=sigma_V, p0=[1000, 2000, 2000, 0], absolute_sigma=False)

y = curve(x, *popt)
ren_res = (curve(t, *popt) - V)/sigma_V
chi_sq = np.sum(ren_res**2)
print("f = ", popt[0], "\nA = ", popt[1],"\nC = ", popt[2], "\nphi = ", popt[3], "\nkappa_sq = ", chi_sq, "\nndof = ", len(t), "\nkappa_sq_norm =", chi_sq / len(t))
print(popt, pcov)

fig = plt.figure('Onda')
fig.add_axes((0.12, 0.3, 0.8, 0.6))
plt.errorbar(t, V, yerr=sigma_V, fmt='.')
plt.plot(x, y)
plt.ylabel('$V_{out}$ [digit]')
plt.xlabel('$t$ [s]')

fig.add_axes((0.12, 0.1, 0.8, 0.2))
plt.errorbar(t, ren_res, yerr=1, fmt='.', capsize=2, markersize=2.5)
y = np.linspace(0.0, 0.0, len(x))
plt.plot(x, y)
plt.ylabel('Residui\nnormalizzati')
plt.xlabel('$t$ [s]')
plt.grid(ls='dashed', which='both', color='gray')
plt.savefig('periodo.png', dpi=800)
plt.show()
```

INGRESSO UN TRENO DI IMPULSI CON DUTY CYCLE VARIABILE CREATO DA ARDUIN IN MODALITÀ PWM; LA FREQUENZA DEL TRENO DI IMPULSI È CIRCA 1 KHZ**

```python
from array import array
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square

n = 10000
fT= 66
f=1000
def run(d):
  x = np.linspace(-2/f,2/f,5000)
  somma = np.zeros_like(x)
  somma += d
  for i in range(1,n+1):
    fk = i * f
    #function = (2/((2*i+1)*np.pi))**2 * np.cos((2*i+1)*x)
    function = 2/(i*np.pi) /np.sqrt(1+(fk/fT)**2) * np.sin(i*np.pi*d) * np.cos(2*np.pi*fk * x - np.arctan(fk/fT))
    somma += function
  return [x,somma]
def ori(d):
  x = np.linspace(-2/f,2/f,5000)
  somma = np.zeros_like(x)
  somma += d
  for i in range(1,n+1):
    fk = i * f
    #function = (2/((2*i+1)*np.pi))**2 * np.cos((2*i+1)*x)
    function = 2/(i*np.pi) * np.sin(i*np.pi*d) * np.cos(2*np.pi*fk * x)
    somma += function
  return [x,somma]

fig = plt.figure('Ricostruzione pinna')

fig.add_axes((0.075, 0.690, 0.4, 0.25))
plt.title('Ricostruzione $\delta=0.14$')
a = run(0.14)
b = ori(0.14)
plt.plot(a[0],a[1], label='Simulazione n=5')
plt.plot(b[0],b[1], label='Simulazione n=5')

fig.add_axes((0.575, 0.690, 0.4, 0.25))
plt.title('Ricostruzione $\delta=0.29$')
a = run(0.29)
b = ori(0.29)
plt.plot(a[0],a[1], label='Simulazione n=5')
plt.plot(b[0],b[1], label='Simulazione n=5')

fig.add_axes((0.075, 0.360, 0.4, 0.25))
plt.title('Ricostruzione $\delta=0.43$')
plt.ylabel("Ampiezza [arb.un.]")
a = run(0.43)
b = ori(0.43)
plt.plot(a[0],a[1], label='Simulazione n=5')
plt.plot(b[0],b[1], label='Simulazione n=5')

fig.add_axes((0.575, 0.360, 0.4, 0.25))
plt.title('Ricostruzione $\delta=0.57$')
plt.ylabel("Ampiezza [arb.un.]")
a = run(0.57)
b = ori(0.57)
plt.plot(a[0],a[1], label='Simulazione n=5')
plt.plot(b[0],b[1], label='Simulazione n=5')

fig.add_axes((0.075, 0.045, 0.4, 0.25))
plt.title('Ricostruzione $\delta=0.71$')
plt.xlabel("Tempo [s]")
a = run(0.71)
b = ori(0.71)
plt.plot(a[0],a[1], label='Simulazione n=5')
plt.plot(b[0],b[1], label='Simulazione n=5')

fig.add_axes((0.575, 0.045, 0.4, 0.25))
plt.title('Ricostruzione $\delta=0.86$')
plt.xlabel("Tempo [s]")
a = run(0.86)
b = ori(0.86)
plt.plot(a[0],a[1], label='Simulazione n=5')
plt.plot(b[0],b[1], label='Simulazione n=5')
plt.show()
```

# PUNTO 8)

**RICOSTRUZIONE DI UNA FORMA D'ONDA QUADRA DI "BASSA FREQUENZA" OSSERVATA CON L'OSCILLOSCOPIO IN ACCOPPIAMENTO AC**

```python
import numpy as np
import matplotlib.pyplot as plt

f = 40
T = f ** -1
f_T = 10
n = 50000
f_k = np.array([k*f for k in range(1, n+1)])
c_k = np.array([2/(k*np.pi) if k % 2 else 0 for k in range(1, n+1)])
G_k = (1 + (f_T / f_k)**2)**(-.5)
phi_k = np.arctan(f_T / f_k)

# DERIVATORE
x = np.linspace(0, 4, 1000)
out = np.zeros_like(x)
for k in range(n):
    out += c_k[k] * G_k[k] *\
        np.sin(2*np.pi*f_k[k] * (x * T) + phi_k[k])

plt.figure("Output del derivatore")
plt.title("Ricostruzione $f=40$ Hz")
plt.plot(x, out)
plt.xlim(0, 4)
plt.tick_params(axis='both', which='both', labelbottom=False, labelleft=False)
plt.savefig('Accoppiamento_AC.png', dpi =800)
plt.show()
```
